```python
import os
import pandas as pd

folder_path = r'C:/Users/PJ455VE/OneDrive - EY/Documents/python training/week 4(pandas)/OneDrive_1_8-16-2023/Question3'
file_extension = '.xlsx'

# Get a list of all Excel files in the folder
excel_files = [file for file in os.listdir(folder_path) if file.endswith(file_extension)]

# Initialize an empty dictionary to store DataFrames and dates
data_frames = {}
file_dates = {}

# Loop through each Excel file and read its data into a DataFrame
for file_name in excel_files:
    full_file_path = os.path.join(folder_path, file_name)
    data_frames[file_name] = pd.read_excel(full_file_path)
    
    # Extract date from file name (assuming it's the first part before "_")
    date_str = file_name.split('_')[0]
    file_dates[file_name] = date_str

# Print the contents of each DataFrame along with its date.
for file_name, df in data_frames.items():
    date_str = file_dates[file_name]
    print(f"Data from {file_name}, Date: {date_str}:\n{df}\n{'='*50}")
```

```python
# Process the DataFrames
for file_name, df in data_frames.items():
    # Extract the date from the dictionary
    date_str = file_dates[file_name]
    
    # Convert the date string to a datetime object with the correct format
    date_obj = pd.to_datetime(date_str, format='%d-%m-%Y')   
    
    # Print the updated DataFrame
    print(f"Data from {file_name}, Date: {date_obj.strftime('%d-%m-%Y')}:\n{df}\n{'='*50}")
```

```python
# Loop through each Excel file and process data
for file_name in excel_files:
    full_file_path = os.path.join(folder_path, file_name)
    df = pd.read_excel(full_file_path)
    
    # Extract date from file name (assuming it's the first part before "_")
    date_str = file_name.split('_')[0]
    date_obj = pd.to_datetime(date_str, format='%d-%m-%Y')
    
    # Add 'Date' column to the DataFrame
    df.insert(0, 'Date', date_obj)
    
    # Melt the DataFrame
    id_vars = ['Date', 'Rating']
    value_vars = [col for col in df.columns if col.endswith('mu')]
    melted_df1 = pd.melt(df, id_vars=id_vars, value_vars=value_vars, var_name='region', value_name='mu')
    
    # Remove "_mu" from the 'region' column values
    melted_df1['region'] = melted_df1['region'].str.replace('_mu', '')
    
    # Append melted DataFrame to the list
    melted_data_frames.append(melted_df1)

# Concatenate all melted DataFrames
final_melted_df1 = pd.concat(melted_data_frames, ignore_index=True)

print(final_melted_df1)
```

```python
# Loop through each Excel file and process data
for file_name in excel_files:
    full_file_path = os.path.join(folder_path, file_name)
    df = pd.read_excel(full_file_path)
    
    # Extract date from file name (assuming it's the first part before "_")
    date_str = file_name.split('_')[0]
    date_obj = pd.to_datetime(date_str, format='%d-%m-%Y')
    
    # Add 'Date' column to the DataFrame
    df.insert(0, 'Date', date_obj)
    
    # Melt the DataFrame
    id_vars = ['Date', 'Rating']
    value_vars = [col for col in df.columns if col.endswith('sigma')]
    melted_df2 = pd.melt(df, id_vars=id_vars, value_vars=value_vars, var_name='region', value_name='sigma')
    
    # Remove "_sigma" from the 'region' column values
    melted_df2['region'] = melted_df2['region'].str.replace('_sigma', '')
    
    # Append melted DataFrame to the list
    melted_data_frames.append(melted_df2)

# Concatenate all melted DataFrames
final_melted_df2 = pd.concat(melted_data_frames, ignore_index=True)

print(final_melted_df2)
```

```python
# Specify columns to use as keys for merging
merge_columns = ['Date', 'Rating','region']

# Merge the two DataFrames based on the specified columns
merged_df = final_melted_df1.merge(final_melted_df2, on=merge_columns, how='outer', suffixes=('mu', 'sigma'))

# Print the merged DataFrame
print(merged_df)
```

```python
# Save the merged DataFrame to an Excel file
output_path = 'merged_data.xlsx'
merged_df.to_excel(output_path, index=False)

print(f"Merged DataFrame saved to '{output_path}'")
```
