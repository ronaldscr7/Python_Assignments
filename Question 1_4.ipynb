```python
import pandas as pd
from datetime import datetime, timedelta

# Step 1: Read data from Excel file
file_path = r'C:/Users/PJ455VE/OneDrive - EY/Documents/python training/week 4(pandas)/OneDrive_1_8-16-2023/Assignment_Pandas.xlsx'
sheet_name = 'Question1'
data = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
data
```

```python
# Task 1: Which region is delivering maximum orders by checking Units column?
max_orders_region = data.groupby('Region')['Units'].sum().idxmax()
max_orders = data.groupby('Region')['Units'].sum().max()

print("Region delivering maximum orders:", max_orders_region)
print("Maximum sum of orders:", max_orders)
```

```python
# Task 2: Check the performance of managers and give ratings
# Define bins and labels for manager ratings
bins = [0, 20, 40, 60, 80, 100]
labels = ['E', 'D', 'C', 'B', 'A']
data['Manager Rating'] = pd.cut(data['Units'], bins=bins, labels=labels)

# Display the managers' performance ratings
print("Manager Performance Ratings:")
print(data[['Manager', 'Units', 'Manager Rating']])
```

```python
# Task 3: Data preprocessing
# Remove special characters from column names
data.columns = data.columns.str.replace('[^a-zA-Z0-9]', '_', regex=True)

# Fill missing values with appropriate values
data['OrderDate'] = pd.to_datetime(data['OrderDate'], errors='coerce')
data['Units'].fillna(0, inplace=True)  # Fill missing Units with 0
data['Sale_amt'].fillna(0, inplace=True)  # Fill missing Sale_amt with 0

# Remove duplicate rows
data.drop_duplicates(inplace=True)

# Display the preprocessed data
print("Preprocessed Data:")
print(data.head())
```

```python
# Task 4: Calculate the number of days from today's date
data['Days_old'] = (datetime.now() - data['OrderDate']).dt.days

# Display the order age in days
print("Order Age in Days:")
print(data[['OrderDate', 'Days_old']])
```

```python
# Task 5: Create a Pivot table and calculate manager-wise total sale and mean sale amount
pivot_table = pd.pivot_table(data, values=['Sale_amt'], index=['Manager'], aggfunc={'Sale_amt': ['sum', 'mean']})

# Display the pivot table
print("Manager-wise Total Sale and Mean Sale Amount:")
print(pivot_table)
```

```python
# Task 6: Create a Pivot table for Manager "Douglas" and calculate total sale amount
douglas_pivot_table = pd.pivot_table(data[data['Manager'] == 'Douglas'],
                                     values=['Sale_amt'],
                                     index=['Region', 'Manager', 'SalesMan'],
                                     aggfunc={'Sale_amt': 'sum'})

# Display the pivot table for Manager "Douglas"
print("Total Sale Amount for Manager Douglas:")
print(douglas_pivot_table)
```

```python
# Task 7: Keep rows with at most two NaNs
filtered_data = data[data.isnull().sum(axis=1) <= 2]

# Display the filtered data
print("Filtered Data with at most two NaNs:")
print(filtered_data)
```

```python
# Task 8: Group by and sort sum of Sale_amt within the group
grouped_data = data.groupby(['Region', 'Manager', 'SalesMan', 'Item'])['Sale_amt'].sum().reset_index()
sorted_grouped_data = grouped_data.sort_values(by='Sale_amt', ascending=False)

# Display the sorted grouped data
print("Sorted Grouped Data by Sum of Sale_amt:")
print(sorted_grouped_data)
```

```python
# Task 9: Split dataframe into groups based on Region and set column values into lists
grouped_lists = data.groupby('Region')[['Manager', 'SalesMan', 'Item']].apply(lambda x: x.values.tolist()).reset_index(name='Grouped Values')

# Display the grouped lists
print("Grouped Lists of Values:")
print(grouped_lists)
```

```python
# Task 10: Check if OrderDate is a business day (weekday) or not
data['Is Business Day'] = data['OrderDate'].dt.weekday < 5

# Display the result
print("Check if OrderDate is a Business Day:")
print(data[['OrderDate', 'Is Business Day']])
```
